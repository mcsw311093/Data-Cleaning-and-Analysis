```python
def duplicate_columns( DataF ):
    DataF = DataF.rename(columns = {None:'new_name'})

    duplicates = [col_name for (col_name,freq) in Counter(DataF.columns).items() if freq > 1] 
    print('{} duplicates '.format(len(duplicates)), duplicates)
    
    for dupli in duplicates: 
        #print(DataF[dupli].values)
        try:
            res = DataF[dupli].values.sum()
            DataF.drop(dupli, axis=1, inplace=True)
            DataF['Total {}'.format(dupli)] = res
            
        except: 
            DataF.drop(dupli, axis=1, inplace=True)
        
        #DataF.drop(dupli, axis=1, inplace=True)
        
    return DataF
```

```python
df_list =[]
for k, v in QorK.items():
            
    for i, vv in enumerate(v):
        print('\nOpening File: ', vv, k)
        with open("{}-{}.csv".format(v[i], k), "rb") as fp:
            df = pickle.load(fp)
            
#         df = pd.read_pickle( "{}-{}.csv".format(v[i], k))
        
        ICD.display(df)
        # Add unique identity to row 
        df['Filing/Acc.No.'] = '{}, {}'.format(k, vv)
        # handle duplicate column name
        df_list.append( duplicate_columns(df) )
        
        #print(df_list[0].shape)
        
        
        abt = pd.concat(df_list, sort=False)
```

```python
def del_empty_col( df ):
    
    for columns in df:
        
        df.replace(0, np.nan, inplace=True)
        
        if df[columns].isna().astype(int).mean() > 0.6:
            df.drop([columns], inplace=True, axis=1)
            
    df.fillna(0)
        
    return df
```
